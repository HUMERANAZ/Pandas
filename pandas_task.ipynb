```python
import pandas as pd
```

```python
df=pd.read_excel(r'AgentPerformance.xlsx',header=None)
```

```python
df.drop([0],inplace=True )
```

```python
df
```

```python
df.columns = df.iloc[0]
```

```python
df.columns
```

```python
df.drop([1],inplace=True)
```

```python
df
```

```python
df.columns = df.columns.str.replace(" ","_")
```

```python
df['Date']=pd.to_datetime(df['Date'])
```

```python
df['Total_Chats']=pd.to_numeric(df['Total_Chats'])
```

```python
df['Average_Rating']=pd.to_numeric(df['Average_Rating'])
df['Total_Feedback']=pd.to_numeric(df['Total_Feedback'])
df['Average_Response_Time']=pd.to_timedelta(df['Average_Response_Time'])
df['Average_Resolution_Time']=pd.to_timedelta(df['Average_Resolution_Time'])
```

```python
df.dtypes
```

```python
df['week']=df['Date'].dt.isocalendar().week
```

```python
df.head(20)
```

```python
l=[]
for i in df['Total_Chats']:
    if i!=0:
        l.append(1)
    else:
        l.append(0)
```

```python
df['working_day']=l
```

```python
df
```

```python
#2 .Total working days for each agents 

df1=df.groupby(['Agent_Name'])['working_day'].sum()
print(df1.head(20))
```

```python
#Find out there avarage rating on weekly basis keep this in a mind that they take two days of leave

working_week_days=df.groupby(['Agent_Name','week'])['working_day'].sum()
```

```python
working_week_days
```

```python
df3=df.groupby(['Agent_Name','week'])['Average_Rating'].sum()/working_week_days
```

```python
df3.head(40)
```

```python
#3. Total query that you hvae taken 
df.groupby(['Agent_Name'])['Total_Chats'].sum()
```

```python
#4. total Feedback that you have received 
df.groupby(['Agent_Name'])['Total_Feedback'].sum()
```

```python
#a agent name who have average rating between 3.5 to 4
df2=df.groupby('Agent_Name')['Average_Rating'].sum()/df1
df4=df2.to_frame(name='Average_rating_month').reset_index()
print(df4)
```

```python
df4[(df4['Average_rating_month']>=3.5) & (df4['Average_rating_month']<=4)][['Agent_Name','Average_rating_month']]
```

```python
#6 . Agent name who have rating lesss then 3.5 
df4[(df4['Average_rating_month']<3.5)][['Agent_Name','Average_rating_month']]
```

```python
#7 . agent name who have rating more then 4.5 
df4[(df4['Average_rating_month']>=4.5)][['Agent_Name','Average_rating_month']]
```

```python
l=[]
for i in (df.groupby('Agent_Name')['Total_Feedback'].sum()):
    l.append(i)
print(l)
```

```python
df4['total_feed']=l
```

```python
df4
```

```python
#8 . how many feedaback agents have received more then 4.5 average

df4[df4['Average_rating_month']>4.5][['Agent_Name','total_feed','Average_rating_month']]
```

```python
#9 . average weekly response time for each agent 
df.groupby(['Agent_Name','week'])['Average_Response_Time'].sum()/working_week_days
```

```python
#10 . average weekely resolution time for each agents 

df.groupby(['Agent_Name','week'])['Average_Resolution_Time'].sum()/working_week_days
```

```python
#11 . list of all agents name 

agent_list=[]
for i in df4['Agent_Name']:
    agent_list.append(i)
```

```python
#12 . percentage of chat on which they have received a feedback 
df['chat_percentage_feedback']=(df['Total_Feedback']/df['Total_Chats']*100)
```

```python
df.groupby('Agent_Name')['chat_percentage_feedback'].mean()
```

```python
#13 . Total contributation hour for each and every agents weekly basis 
df['daily_hours']=df['Total_Chats']*df['Average_Resolution_Time']
```

```python
df.groupby(['Agent_Name','week'])['daily_hours'].sum()
```
